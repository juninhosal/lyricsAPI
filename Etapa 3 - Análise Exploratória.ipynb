Análise de Distribuiçao por Estilo Musical

```
    x1 = df_lyrics.loc[df_lyrics['band'] == "massacration-musicas", 'band']
    x2 = df_lyrics.loc[df_lyrics['band'] == "metallica", 'band']
    x3 = df_lyrics.loc[df_lyrics['band'] == "ikimonogakari", 'band']
    x4 = df_lyrics.loc[df_lyrics['band'] == "lisa", 'band']
    x5 = df_lyrics.loc[df_lyrics['band'] == "o-teatro-magico", 'band']
    x6 = df_lyrics.loc[df_lyrics['band'] == "guilherme-de-sa", 'band']

    trace1 = go.Histogram(x=x1, name='Heavy Metal', opacity=0.75)
    trace2 = go.Histogram(x=x2, name='Heavy Metal', opacity=0.75)
    trace3 = go.Histogram(x=x3, name='J-Rock', opacity=0.75)
    trace4 = go.Histogram(x=x4, name='J-Rock', opacity=0.75)
    trace5 = go.Histogram(x=x5, name='Alternativo', opacity=0.75)
    trace6 = go.Histogram(x=x6, name='Alternativo', opacity=0.75)

    data = [trace1, trace2, trace3, trace4, trace5, trace6]
    layout = go.Layout(barmode='overlay', title='Análise de Distribuição por Estilo Musical')
    fig = go.Figure(data=data, layout=layout)

    iplot(fig, filename='overlaid histogram')
```
